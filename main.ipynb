```python
from tabulate import tabulate
from statistics import stdev
import numpy

IgboDocName = (r'file1.txt') 
YorubaDocName = r"file2.txt"

validLetters = set(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'ị', 'j', 'k', 'l', 'm', 'n', 'o', 'ọ', 'p', 'q', 'r', 's', 'ṣ', 't', 'u', 'ụ', 'v', 'w', 'x', 'y', 'z']) 
 

#IGBO character count
letterCount = {} 
letterCount.update([(x,0) for x in validLetters])

with open(IgboDocName, "r", encoding="utf8") as wordFile: 
    for letter in wordFile.read().lower(): 
        if letter in validLetters: 
            letterCount[letter] += 1 

table_headers = ["Character", "Frequency"]
lang1_data_list = [(char, count) for char, count in letterCount.items()]

print(tabulate(sorted(lang1_data_list), headers=table_headers, tablefmt="fancy_grid"))


#Yoruba character count
letterCount = {} 
letterCount.update([(x,0) for x in validLetters])

with open(YorubaDocName, "r", encoding="utf8") as wordFile: 
    for letter in wordFile.read().lower(): 
        if letter in validLetters: 
            letterCount[letter] += 1 

table_headers = ["Character", "Frequency"]
lang2_data_list = [(char, count) for char, count in letterCount.items()]

print(tabulate(sorted(lang2_data_list), headers=table_headers, tablefmt="fancy_grid"))


lang1_frequency_list = [item[1] for item in lang1_data_list]
lang2_frequency_list = [item[1] for item in lang2_data_list]

x_stdev = stdev(lang1_frequency_list)
y_stdev = stdev(lang2_frequency_list)

correlation_coeff = numpy.corrcoef(lang1_frequency_list, lang2_frequency_list)
print(correlation_coeff)

# for letter in sorted(validLetters): 
#     print ("Character {} Count {}".format(letter,letterCount[letter])) 
```
